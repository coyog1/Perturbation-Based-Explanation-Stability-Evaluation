```python
def get_explanation(sample_index: int,
                    cp: int = 6,
                    model_type: str = "class",
                    distance: str = "dtw"):
    """
    Returns a dict with:
      - sample: 1D numpy array of the time series
      - class_idx: int, predicted class
      - importances: np.ndarray of shape (n_segments,) or (n_classes,n_segments)
      - boundaries: list of segment boundary indices
    """
    raw = x_test[sample_index]
    sample = raw.reshape(-1)

    # predicted class
    probs = model.predict(sample[np.newaxis, :, np.newaxis])
    class_idx = int(np.argmax(probs))

    # call LIMESegment
    
    importances, boundaries = LIMESegment(
        sample,
        model,
        model_type=model_type,
        distance=distance,
        cp=cp
    )
    print("Expl. made")

    return {
        "sample": sample,
        "class_idx": class_idx,
        "importances": importances,
        "boundaries": boundaries
    }
```

```python
def plot_explanation(explanation: dict,
                     title: str = None,
                     figsize=(14, 6)):
    """
    Plots the time series with shaded segments and importance annotations.
    Expects explanation dict from get_explanation().
    """
    sample = explanation["sample"]
    importances = explanation["importances"]
    boundaries = explanation["boundaries"]
    class_idx = explanation["class_idx"]

    T = sample.shape[0]
    y_min, y_max = sample.min(), sample.max()
    y_range = y_max - y_min

    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

    plt.figure(figsize=figsize)
    plt.plot(sample, color='black', linewidth=1.5, label="Time Series")

    # shaded segments + annotations
    for i in range(len(boundaries)-1):
        start, end = boundaries[i], boundaries[i+1] if boundaries[i+1]>=0 else T
        c = colors[i % len(colors)]
        plt.axvspan(start, end, color=c, alpha=0.25)
        imp = (importances[class_idx][i]
               if importances.ndim==2 else importances[i])
        mid = (start+end)/2
        ann_y = sample[start:end].max() + 0.1*y_range
        plt.text(mid, ann_y, f"{imp:.4f}",
                 ha='center', va='bottom',
                 fontsize=11, fontweight='bold', color=c)

    # legend
    from matplotlib.patches import Patch
    patches = [Patch(color=colors[i % len(colors)],
                     label=f"Segment {i+1}")
               for i in range(len(boundaries)-1)]
    plt.legend(handles=patches, loc="upper right")

    plt.title(title or f"LIMESegment Explanation (class {class_idx})",
              fontsize=16, fontweight='bold')
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.ylim(y_min - 0.1*y_range, y_max + 0.2*y_range)
    plt.tight_layout()
    plt.show()
```

```python
def build_attribution_map(explanation: dict) -> np.ndarray:
    """
    Given explanation dict, returns 1D attribution map of length T.
    """
    sample = explanation["sample"]
    importances = explanation["importances"]
    boundaries = explanation["boundaries"]
    class_idx = explanation["class_idx"]

    T = sample.shape[0]
    attr = np.zeros(T, dtype=float)
    seg_imp = (importances[class_idx]
               if importances.ndim==2 else importances)

    # fix final boundary
    bnds = list(boundaries)
    if bnds[-1] < 0:
        bnds[-1] = T

    for i in range(len(bnds)-1):
        start = max(0, bnds[i])
        end = min(T, bnds[i+1])
        if start < end:
            attr[start:end] += seg_imp[i]

    return attr
```

```python
def plot_attribution_map(attr_map: np.ndarray,
                         title: str = None,
                         figsize=(12, 4),
                         color='tab:blue'):
    """
    Plots a 1D attribution map including negative values.
    """
    T = len(attr_map)
    x = np.arange(T)

    plt.figure(figsize=figsize)
    # draw zero line
    plt.axhline(0, color='gray', linewidth=1)

    # fill positives
    plt.fill_between(x, 0, attr_map, where=attr_map >= 0, color=color, alpha=0.3, interpolate=True)
    # fill negatives
    plt.fill_between(x, 0, attr_map, where=attr_map < 0, color=color, alpha=0.3, interpolate=True)

    plt.plot(x, attr_map, color=color, linewidth=1.5)
    plt.title(title or "Attribution Map", fontsize=14, fontweight='bold')
    plt.xlabel("Time Step")
    plt.ylabel("Attribution")
    # expand y‐limits to include negatives and positives
    y_min, y_max = attr_map.min(), attr_map.max()
    plt.ylim(y_min * 1.1 if y_min < 0 else 0,
             y_max * 1.1 if y_max > 0 else 0)
    plt.tight_layout()
    plt.show()

def predict_series(sample: np.ndarray) -> int:
    """Return the predicted class for a single 1-D series."""
    probs = model.predict(sample.reshape(1, -1, 1))
    return int(np.argmax(probs, axis=1)[0])
```

##### Phase Shift

```python
def phase_shift(idx: int, phi: float) -> np.ndarray: # purely grabs X_test[idx] and then calls the math function
    sample = x_test[idx].reshape(-1)
    return phase_shift_calc(sample, phi)

def phase_shift_calc(sample: np.ndarray, phi: float) -> np.ndarray: # purely does the math on the array
    analytic = hilbert(sample)
    return np.real(analytic * np.exp(1j * phi))

def plot_phase_shift(idx: int = None, phi: float = None, figsize: tuple = (10, 4), original_sample: np.ndarray = None, shifted_sample: np.ndarray = None):
    
    if original_sample is not None and shifted_sample is not None:
        # Use pre-computed samples
        original = original_sample
        shifted = shifted_sample
    else:
        # Compute on-the-fly (your original approach)
        original = x_test[idx].reshape(-1)
        shifted = phase_shift(idx, phi)
    
    T = len(original)
    t = np.arange(T)
    plt.figure(figsize=figsize)
    plt.plot(t, original, label="Original", color="black", lw=1.5)
    plt.plot(t, shifted, label=f"Shifted (ϕ={phi:.2f} rad)", ls="--", lw=1.5)
    title = "Original vs. Phase-Shifted"
    if idx is not None:
        title += f" (sample {idx})"
    plt.title(title, fontsize=14, fontweight="bold")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()
```

##### Frequency Mask

```python
def frequency_mask(idx: int, freq_mask_param: int, num_freq_masks: int) -> np.ndarray:
    sample = x_test[idx].reshape(-1)
    return frequency_mask_calc(sample, freq_mask_param, num_freq_masks)

def frequency_mask_calc(sample: np.ndarray, freq_mask_param: int, num_freq_masks: int = 1, mask_value: float = 0.0) -> np.ndarray:
    """
    freq_mask_param : Maximum width (in frequency bins) of each mask.
    num_freq_masks : Number of independent frequency masks to apply.
    mask_value : Value to assign to masked frequencies.
    """
    # Real‐valued FFT (keeps non‐negative freqs)
    spec = np.fft.rfft(sample)
    n_freq_bins = spec.shape[0]

    for _ in range(num_freq_masks):
        # random start bin for this mask
        f0 = np.random.randint(0, max(1, n_freq_bins - freq_mask_param))
        f1 = f0 + freq_mask_param
        spec[f0:f1] = mask_value

    # invert back to time domain
    return np.fft.irfft(spec, n=sample.shape[0])


def plot_frequency_mask(idx: int, freq_mask_param: int, num_freq_masks: int, figsize: tuple = (10, 4)):
    original = x_test[idx].reshape(-1)
    masked   = frequency_mask(idx, freq_mask_param, num_freq_masks)

    T = original.shape[0]
    t = np.arange(T)

    plt.figure(figsize=figsize)
    plt.plot(t, original, label="Original", color="black", lw=1.5)
    plt.plot(t, masked,   label=f"FreqMask(F={freq_mask_param}, nM={num_freq_masks})",
             ls="--", lw=1.5)
    plt.title(f"Frequency Masking (sample {idx})", fontsize=14, fontweight="bold")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()
```

##### Value Mask

```python
def value_mask(idx: int, mask_width: int, num_masks: int = 1, mask_value: float = 0.0) -> np.ndarray:
    sample = x_test[idx].reshape(-1)
    return value_mask_calc(sample, mask_width, num_masks, mask_value)

def value_mask_calc(sample: np.ndarray, mask_width: int, num_masks: int = 1, mask_value: float = 0.0) -> np.ndarray:
    """mask_width : Maximum length (in time steps) of each mask"""
    T = sample.shape[0]
    masked = sample.copy()
    for _ in range(num_masks):
        start = np.random.randint(0, max(1, T - mask_width + 1))
        end = start + mask_width
        masked[start:end] = mask_value
    return masked

def plot_value_mask(idx: int, mask_width: int, num_masks: int = 1, mask_value: float = 0.0, figsize: tuple = (10, 4)):
    original = x_test[idx].reshape(-1)
    masked = value_mask(idx, mask_width, num_masks, mask_value)
    T = original.shape[0]
    t = np.arange(T)

    plt.figure(figsize=figsize)
    plt.plot(t, original, label="Original", color="black", linewidth=1.5)
    plt.plot(t, masked,
             label=f"ValueMask(w={mask_width}, n={num_masks})",
             linestyle="--", linewidth=1.5)
    plt.title(f"Value Masking (sample {idx})", fontsize=14, fontweight="bold")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()
```

##### Amplitude Scale

```python
def amplitude_scale(idx: int, scale_factor: float) -> np.ndarray:
    sample = x_test[idx].reshape(-1)
    return amplitude_scale_calc(sample, scale_factor)

def amplitude_scale_calc(sample: np.ndarray,
                          scale_factor: float) -> np.ndarray:
    """Multiplicative scaling factor. 
        - scale_factor > 1.0: increases amplitude
        - scale_factor < 1.0: decreases amplitude  
        - scale_factor = 1.0: no change"""
    return sample * scale_factor

def plot_amplitude_scale(idx: int, scale_factor: float, figsize: tuple = (10, 4)):
    original = x_test[idx].reshape(-1)
    scaled = amplitude_scale(idx, scale_factor)
    T = original.shape[0]
    t = np.arange(T)

    plt.figure(figsize=figsize)
    plt.plot(t, original, label="Original", color="black", linewidth=1.5)
    plt.plot(t, scaled,
             label=f"Scaled (×{scale_factor})",
             linestyle="--", linewidth=1.5)
    plt.title(f"Amplitude Scaling (sample {idx})", fontsize=14, fontweight="bold")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()
```

##### Contiguous Mask

```python
def contiguous_mask(idx: int, mask_width: int, mask_value: float = 0.0) -> np.ndarray:
    sample = x_test[idx].reshape(-1)
    return contiguous_mask_calc(sample, mask_width, mask_value)

def contiguous_mask_calc(sample: np.ndarray, mask_width: int, mask_value: float = 0.0) -> np.ndarray:
    """mask_width : Width of the contiguous mask in time steps.
       mask_value : Value to assign to masked time steps."""
    T = sample.shape[0]
    masked = sample.copy()
    
    # Randomly select starting position for the contiguous mask
    start = np.random.randint(0, max(1, T - mask_width + 1))
    end = start + mask_width
    
    # Apply the contiguous mask
    masked[start:end] = mask_value
    
    return masked

def plot_contiguous_mask(idx: int, mask_width: int, mask_value: float = 0.0, figsize: tuple = (10, 4)):
    original = x_test[idx].reshape(-1)
    masked = contiguous_mask(idx, mask_width, mask_value)
    T = original.shape[0]
    t = np.arange(T)

    plt.figure(figsize=figsize)
    plt.plot(t, original, label="Original", color="black", linewidth=1.5)
    plt.plot(t, masked,
             label=f"ContiguousMask(w={mask_width})",
             linestyle="--", linewidth=1.5)
    plt.title(f"Contiguous Masking (sample {idx})", fontsize=14, fontweight="bold")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()
```

##### Unified Perturbation Plotting

```python
def plot_perturbation_comparison(idx: int, perturbation_type: str,
    # Phase shift parameters
    phi: float = None,
    # Frequency masking parameters  
    freq_mask_param: int = None, num_freq_masks: int = 1,
    # Value masking parameters
    mask_width: int = None, num_value_masks: int = 1, mask_value: float = 0.0,
    # Amplitude scaling parameters
    scale_factor: float = None,
    # Contiguous masking parameters
    contiguous_mask_width: int = None,
                                 
    # Universal plotting parameters
    figsize: tuple = (10, 4), title: str = None, original_color: str = "black", perturbed_color: str = "red", original_linewidth: float = 1.5,
    perturbed_linewidth: float = 1.5, perturbed_linestyle: str = "--", alpha: float = 0.8, show_grid: bool = False, legend_loc: str = "upper right"
):
   
    # Extract original sample
    original = x_test[idx].reshape(-1)
    
    # Apply the specified perturbation
    if perturbation_type == "phase_shift":
        if phi is None:
            raise ValueError("phi parameter is required for phase_shift perturbation")
        perturbed = phase_shift(idx, phi)
        param_str = f"ϕ={phi:.2f} rad"
        
    elif perturbation_type == "frequency_mask":
        if freq_mask_param is None:
            raise ValueError("freq_mask_param is required for frequency_mask perturbation")
        perturbed = frequency_mask(idx, freq_mask_param, num_freq_masks)
        param_str = f"F={freq_mask_param}, nM={num_freq_masks}"
        
    elif perturbation_type == "value_mask":
        if mask_width is None:
            raise ValueError("mask_width is required for value_mask perturbation")
        perturbed = value_mask(idx, mask_width, num_value_masks, mask_value)
        param_str = f"w={mask_width}, n={num_value_masks}"
        
    elif perturbation_type == "amplitude_scale":
        if scale_factor is None:
            raise ValueError("scale_factor is required for amplitude_scale perturbation")
        perturbed = amplitude_scale(idx, scale_factor)
        param_str = f"×{scale_factor}"
        
    elif perturbation_type == "contiguous_mask":
        if contiguous_mask_width is None:
            raise ValueError("contiguous_mask_width is required for contiguous_mask perturbation")
        perturbed = contiguous_mask(idx, contiguous_mask_width, mask_value)
        param_str = f"w={contiguous_mask_width}"
        
    else:
        raise ValueError(f"Unknown perturbation_type: {perturbation_type}")
    
    # Create the plot
    T = len(original)
    t = np.arange(T)
    
    plt.figure(figsize=figsize)
    
    # Plot original and perturbed series
    plt.plot(t, original, 
             label="Original", 
             color=original_color, 
             linewidth=original_linewidth, 
             alpha=alpha)
    
    plt.plot(t, perturbed, 
             label=f"{perturbation_type.replace('_', ' ').title()} ({param_str})", 
             color=perturbed_color, 
             linewidth=perturbed_linewidth, 
             linestyle=perturbed_linestyle, 
             alpha=alpha)
    
    # Customize plot
    if title is None:
        title = f"{perturbation_type.replace('_', ' ').title()} Perturbation (sample {idx})"
    
    plt.title(title, fontsize=14, fontweight="bold")
    plt.xlabel("Time Step", fontsize=12)
    plt.ylabel("Value", fontsize=12)
    plt.legend(loc=legend_loc)
    
    if show_grid:
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

##### Unified Perturbation Calculation

```python
def apply_perturbation(idx: int, perturbation_type: str,
                       # phase shift
                       phi: float = None,
                       # frequency mask
                       freq_mask_param: int = None, num_freq_masks: int = 1,
                       # value mask
                       mask_width: int = None, num_value_masks: int = 1, mask_value: float = 0.0,
                       # amplitude scale
                       scale_factor: float = None,
                       # contiguous mask
                       contiguous_mask_width: int = None ) -> np.ndarray:
    """
    Return the perturbed time series for X_test[idx], according to perturbation_type.
    """
    sample = x_test[idx].reshape(-1)
    if perturbation_type == "phase_shift":
        if phi is None:
            raise ValueError("phi is required for phase_shift")
        return phase_shift(idx, phi)

    elif perturbation_type == "frequency_mask":
        if freq_mask_param is None:
            raise ValueError("freq_mask_param is required for frequency_mask")
        return frequency_mask(idx, freq_mask_param, num_freq_masks)

    elif perturbation_type == "value_mask":
        if mask_width is None:
            raise ValueError("mask_width is required for value_mask")
        return value_mask(idx, mask_width, num_value_masks, mask_value)

    elif perturbation_type == "amplitude_scale":
        if scale_factor is None:
            raise ValueError("scale_factor is required for amplitude_scale")
        return amplitude_scale(idx, scale_factor)

    elif perturbation_type == "contiguous_mask":
        if contiguous_mask_width is None:
            raise ValueError("contiguous_mask_width is required for contiguous_mask")
        return contiguous_mask(idx, contiguous_mask_width, mask_value)

    else:
        raise ValueError(f"Unknown perturbation_type: {perturbation_type}")
```

### Perturbation Validity Check

##### Class Flip Check

```python
def class_flip_check(idx: int, perturbation_type: str, **perturb_kwargs) -> dict:
    """Check whether a perturbation flips the model’s predicted class."""
    # 1) original prediction
    original = x_test[idx].reshape(-1)
    orig_pred = predict_series(original)
    
    # 2) build perturbed series
    perturbed = apply_perturbation(idx, perturbation_type, **perturb_kwargs )
    
    # 3) perturbed prediction
    pert_pred = predict_series(perturbed)
    
    return {
        "original_pred": orig_pred,
        "perturbed_pred": pert_pred,
        "flipped": orig_pred != pert_pred
    }
```

```python
import numpy as np
import warnings
from scipy.spatial.distance import mahalanobis
from sklearn.covariance import EmpiricalCovariance
from sklearn.model_selection import train_test_split

class CorrectedTimeSeriesOODDetector:
    """
    Properly calibrated OOD detector that avoids training-test contamination.
    """
    
    def __init__(self, x_train: np.ndarray, x_test: np.ndarray, method: str = "hybrid"):
        """
        Initialize OOD detector with proper calibration approach.
        
        Parameters
        ----------
        x_train : np.ndarray
            Training data for fitting the detector.
        x_test : np.ndarray  
            Test data for proper threshold calibration.
        method : str
            Detection method: "mahalanobis", "statistical", or "hybrid"
        """
        self.x_train = x_train.reshape(x_train.shape[0], -1)
        self.x_test = x_test.reshape(x_test.shape[0], -1) 
        self.method = method
        self.is_fitted = False
        
        # Create calibration set from combined training and test data
        self._create_calibration_set()
        self.fit()
    
    def _create_calibration_set(self):
        """
        Create proper calibration set by combining training and test data.
        This avoids the training-test contamination issue.
        """
        # Combine training and test data for calibration
        combined_data = np.vstack([self.x_train, self.x_test])
        
        # Split combined data into fitting set and calibration set
        self.fit_data, self.calibration_data = train_test_split(
            combined_data, test_size=0.3, random_state=42
        )
        
        print(f"Created calibration set: {len(self.fit_data)} fit samples, {len(self.calibration_data)} calibration samples")
    
    def fit(self):
        """Fit the OOD detector using the fitting set."""
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            
            if self.method in ["mahalanobis", "hybrid"]:
                self._fit_mahalanobis()
            if self.method in ["statistical", "hybrid"]:
                self._fit_statistical()
        
        self.is_fitted = True
    
    def _fit_mahalanobis(self):
        """Fit Mahalanobis distance detector with proper calibration."""
        # Compute statistics from fitting set
        self.train_mean = np.mean(self.fit_data, axis=0)
        
        # Use shrinkage estimator for better covariance estimation
        cov_estimator = EmpiricalCovariance(assume_centered=False)
        cov_estimator.fit(self.fit_data)
        self.train_cov_inv = cov_estimator.precision_
        
        # CRITICAL: Calibrate threshold using calibration set, not training set
        calibration_distances = [
            mahalanobis(sample, self.train_mean, self.train_cov_inv)
            for sample in self.calibration_data
        ]
        
        # Use more conservative threshold (85th percentile instead of 95th)
        self.mahal_threshold = np.percentile(calibration_distances, 85)
        print(f"Mahalanobis threshold calibrated: {self.mahal_threshold:.4f}")
    
    def _fit_statistical(self):
        """Fit statistical detector with proper calibration."""
        # Compute statistics from fitting set
        fit_means = np.mean(self.fit_data, axis=1)
        fit_stds = np.std(self.fit_data, axis=1)
        
        self.fit_mean_dist = (np.mean(fit_means), np.std(fit_means))
        self.fit_std_dist = (np.mean(fit_stds), np.std(fit_stds))
        
        # Calibrate using calibration set
        cal_means = np.mean(self.calibration_data, axis=1)
        cal_stds = np.std(self.calibration_data, axis=1)
        
        mean_z_scores = np.abs(cal_means - self.fit_mean_dist[0]) / self.fit_mean_dist[1]
        std_z_scores = np.abs(cal_stds - self.fit_std_dist[0]) / self.fit_std_dist[1]
        combined_scores = np.maximum(mean_z_scores, std_z_scores)
        
        # Use 85th percentile threshold
        self.stat_threshold = np.percentile(combined_scores, 85)
        print(f"Statistical threshold calibrated: {self.stat_threshold:.4f}")
    
    def detect_ood(self, sample: np.ndarray) -> dict:
        """
        Detect if a sample is OOD using properly calibrated thresholds.
        """
        if not self.is_fitted:
            raise ValueError("Detector must be fitted before use")
        
        flat_sample = sample.reshape(-1)
        
        if self.method == "mahalanobis":
            return self._detect_mahalanobis(flat_sample)
        elif self.method == "statistical":
            return self._detect_statistical(flat_sample)
        elif self.method == "hybrid":
            return self._detect_hybrid(flat_sample)
    
    def _detect_mahalanobis(self, sample: np.ndarray) -> dict:
        """Mahalanobis detection with proper calibration."""
        distance = mahalanobis(sample, self.train_mean, self.train_cov_inv)
        is_ood = distance > self.mahal_threshold
        
        return {
            "is_ood": bool(is_ood),
            "score": float(distance),
            "threshold": float(self.mahal_threshold),
            "method": "mahalanobis_calibrated"
        }
    
    def _detect_statistical(self, sample: np.ndarray) -> dict:
        """Statistical detection with proper calibration."""
        sample_mean = np.mean(sample)
        sample_std = np.std(sample)
        
        mean_z = abs(sample_mean - self.fit_mean_dist[0]) / self.fit_mean_dist[1]
        std_z = abs(sample_std - self.fit_std_dist[0]) / self.fit_std_dist[1]
        combined_score = max(mean_z, std_z)
        
        is_ood = combined_score > self.stat_threshold
        
        return {
            "is_ood": bool(is_ood),
            "score": float(combined_score),
            "threshold": float(self.stat_threshold),
            "method": "statistical_calibrated"
        }
    
    def _detect_hybrid(self, sample: np.ndarray) -> dict:
        """Hybrid detection combining both methods."""
        mahal_result = self._detect_mahalanobis(sample)
        stat_result = self._detect_statistical(sample)
        
        # Conservative approach: require both methods to agree
        is_ood = mahal_result["is_ood"] and stat_result["is_ood"]
        
        return {
            "is_ood": bool(is_ood),
            "score": (mahal_result["score"] / mahal_result["threshold"] + 
                     stat_result["score"] / stat_result["threshold"]) / 2,
            "threshold": 1.0,
            "method": "hybrid_calibrated",
            "details": {
                "mahalanobis": mahal_result,
                "statistical": stat_result
            }
        }

# Corrected usage function
def create_corrected_ood_detector(x_train: np.ndarray, 
                                 x_test: np.ndarray,
                                 method: str = "hybrid") -> CorrectedTimeSeriesOODDetector:
    """
    Create properly calibrated OOD detector that avoids test contamination.
    """
    return CorrectedTimeSeriesOODDetector(x_train, x_test, method=method)
```

### Metrics

```python
def circular_cross_correlation(x: np.ndarray, y: np.ndarray) -> dict:
    # Ensure the same length
    assert len(x) == len(y), "Attribution maps must have same length"
    
    # Periodic cross-correlation using FFT
    from numpy.fft import fft, ifft
    correlation = ifft(fft(x) * fft(y).conj()).real
    
    # Find maximum correlation and corresponding lag
    max_idx = np.argmax(np.abs(correlation))
    norm = np.linalg.norm(x)*np.linalg.norm(y)
    max_corr = correlation[max_idx]/ norm
    
    # Convert index to lag (considering circular nature)
    lag = max_idx if max_idx <= len(x)//2 else max_idx - len(x)
    
    return {
        "max_correlation": float(max_corr),
        "lag_at_max": int(lag),
        "full_correlation": correlation
    }

def cosine_similarity_metric(x: np.ndarray, y: np.ndarray) -> float:
    """Returns similarity in [0, 1] where 1 means identical direction."""
    # Handle zero vectors
    if np.allclose(x, 0) and np.allclose(y, 0):
        return 1.0
    if np.allclose(x, 0) or np.allclose(y, 0):
        return 0.0

    return dot(x, y)/(norm(x)*norm(y))

def spectral_correlation(x: np.ndarray, y: np.ndarray) -> dict:
    """Compute correlation in frequency domain between attribution maps. Returns both magnitude and phase correlations."""
    # Compute FFT of both signals
    X_fft = np.fft.fft(x)
    Y_fft = np.fft.fft(y)
    
    # Magnitude correlation
    X_mag = np.abs(X_fft)
    Y_mag = np.abs(Y_fft)
    mag_corr = np.corrcoef(X_mag, Y_mag)[0, 1] if not np.allclose(X_mag, Y_mag) else 1.0
    
    # Phase correlation  
    X_phase = np.angle(X_fft)
    Y_phase = np.angle(Y_fft)
    phase_corr = np.corrcoef(X_phase, Y_phase)[0, 1] if not np.allclose(X_phase, Y_phase) else 1.0
    
    # Handle NaN cases
    mag_corr = 0.0 if np.isnan(mag_corr) else mag_corr
    phase_corr = 0.0 if np.isnan(phase_corr) else phase_corr
    
    return {
        "magnitude_correlation": float(mag_corr),
        "phase_correlation": float(phase_corr),
        "combined_spectral_score": float(0.7 * mag_corr + 0.3 * phase_corr)
    }

def earth_mover_distance(x: np.ndarray, y: np.ndarray) -> float:
    """Compute Earth Mover's Distance (Wasserstein-1) between attribution maps. Treats attribution maps as 1D distributions."""
    # Create position arrays
    positions = np.arange(len(x))
    
    # Use absolute values and normalize to create valid distributions
    x_pos = np.abs(x)
    y_pos = np.abs(y)
    
    # Add small epsilon to avoid zero distributions
    epsilon = 1e-10
    x_pos += epsilon
    y_pos += epsilon
    
    # Normalize to probability distributions
    x_pos = x_pos / np.sum(x_pos)
    y_pos = y_pos / np.sum(y_pos)
    
    return wasserstein_distance(positions, positions, x_pos, y_pos)

def fourier_distance(x: np.ndarray, y: np.ndarray, n_coeffs: int = None) -> float:
    """n_coeffs: Number of Fourier coefficients to use. If None, uses all."""
    # Compute FFT
    X_fft = np.fft.fft(x)
    Y_fft = np.fft.fft(y)
    
    # Use specified number of coefficients
    if n_coeffs is None:
        n_coeffs = len(x) // 2 + 1
    
    X_coeffs = X_fft[:n_coeffs]
    Y_coeffs = Y_fft[:n_coeffs]
    
    # Euclidean distance between coefficients
    return float(np.linalg.norm(X_coeffs - Y_coeffs))

def jensen_shannon_divergence(x: np.ndarray, y: np.ndarray) -> float:
    """
    Compute Jensen-Shannon divergence between attribution maps.
    
    Converts attribution maps to probability distributions.
    """
    # Convert to non-negative distributions
    x_pos = np.abs(x) + 1e-10
    y_pos = np.abs(y) + 1e-10
    
    # Normalize to probability distributions
    x_prob = x_pos / np.sum(x_pos)
    y_prob = y_pos / np.sum(y_pos)
    
    return jensenshannon(x_prob, y_prob)

def pearson_correlation(x: np.ndarray, y: np.ndarray) -> float:
    """Compute Pearson correlation coefficient."""
    corr, _ = stats.pearsonr(x, y)
    return float(corr) if not np.isnan(corr) else 0.0

def spearman_correlation(x: np.ndarray, y: np.ndarray) -> float:
    """Compute Spearman rank correlation coefficient."""
    corr, _ = stats.spearmanr(x, y)
    return float(corr) if not np.isnan(corr) else 0.0

def l1_distance(x: np.ndarray, y: np.ndarray) -> float:
    """Compute L1 (Manhattan) distance."""
    return float(mean_absolute_error(x, y))

def l2_distance(x: np.ndarray, y: np.ndarray) -> float:
    """Compute L2 (Euclidean) distance."""
    return float(np.sqrt(mean_squared_error(x, y)))

def compute_all_attribution_metrics(x: np.ndarray, y: np.ndarray) -> dict: 
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        
        # Circular cross-correlation
        circ_corr = circular_cross_correlation(x, y)
        
        # Spectral analysis
        spectral = spectral_correlation(x, y)
        
        metrics = {
            # Cross-correlation metrics
            "circular_cross_correlation": circ_corr["max_correlation"],
            "optimal_lag": circ_corr["lag_at_max"],
            
            # Similarity metrics
            "cosine_similarity": cosine_similarity_metric(x, y),
            "pearson_correlation": pearson_correlation(x, y), 
            "spearman_correlation": spearman_correlation(x, y),
            
            # Spectral metrics
            "spectral_magnitude_correlation": spectral["magnitude_correlation"],
            "spectral_phase_correlation": spectral["phase_correlation"],
            "combined_spectral_score": spectral["combined_spectral_score"],
            
            # Distance metrics
            "earth_mover_distance": earth_mover_distance(x, y),
            "fourier_distance": fourier_distance(x, y),
            "jensen_shannon_divergence": jensen_shannon_divergence(x, y),
            "l1_distance": l1_distance(x, y),
            "l2_distance": l2_distance(x, y)
        }
    
    return metrics
```

```python
def plot_attribution_comparison(original_attr: np.ndarray,
                               perturbed_attr: np.ndarray,
                               metrics: dict,
                               original_label: str = "Original",
                               perturbed_label: str = "Perturbed",
                               sample_idx: int = None,
                               figsize: tuple = (15, 8)) -> None:
    """
    Plot side-by-side attribution map comparison with metrics.
    
    Parameters
    ----------
    original_attr, perturbed_attr : np.ndarray
        Attribution maps to compare and visualize.
    metrics : dict
        Computed metrics from compute_all_attribution_metrics.
    original_label, perturbed_label : str
        Labels for the attribution maps.
    sample_idx : int, optional
        Sample index for title.
    figsize : tuple
        Figure size.
    """
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=figsize)
    
    T = len(original_attr)
    t = np.arange(T)
    
    # Plot original attribution map
    ax1.plot(t, original_attr, color="blue", linewidth=1.5)
    ax1.fill_between(t, 0, original_attr, color="blue", alpha=0.3)
    ax1.axhline(0, color='gray', linewidth=1, alpha=0.5)
    ax1.set_title(f"{original_label} Attribution Map", fontsize=12, fontweight="bold")
    ax1.set_xlabel("Time Step")
    ax1.set_ylabel("Attribution")
    ax1.grid(True, alpha=0.3)
    
    # Plot perturbed attribution map
    ax2.plot(t, perturbed_attr, color="red", linewidth=1.5)
    ax2.fill_between(t, 0, perturbed_attr, color="red", alpha=0.3)
    ax2.axhline(0, color='gray', linewidth=1, alpha=0.5)
    ax2.set_title(f"{perturbed_label} Attribution Map", fontsize=12, fontweight="bold")
    ax2.set_xlabel("Time Step")
    ax2.set_ylabel("Attribution")
    ax2.grid(True, alpha=0.3)
    
    # Display metrics as text
    ax3.axis('off')
    ax3.text(0.05, 0.95, "Attribution Comparison Metrics:", 
             fontsize=14, fontweight="bold", transform=ax3.transAxes, va='top')
    
    y_pos = 0.85
    
    # Group metrics by category for better readability (FIXED: removed colon from fmt)
    similarity_metrics = [
        ("Cosine Similarity", "cosine_similarity", ".4f"),
        ("Pearson Correlation", "pearson_correlation", ".4f"),
        ("Spearman Correlation", "spearman_correlation", ".4f"),
        ("Circular Cross-Correlation", "circular_cross_correlation", ".4f"),
    ]
    
    spectral_metrics = [
        ("Spectral Magnitude Corr.", "spectral_magnitude_correlation", ".4f"),
        ("Spectral Phase Corr.", "spectral_phase_correlation", ".4f"),
        ("Combined Spectral Score", "combined_spectral_score", ".4f"),
    ]
    
    distance_metrics = [
        ("Earth Mover's Distance", "earth_mover_distance", ".4f"),
        ("Fourier Distance", "fourier_distance", ".4f"),
        ("Jensen-Shannon Divergence", "jensen_shannon_divergence", ".4f"),
        ("L1 Distance", "l1_distance", ".4f"),
        ("L2 Distance", "l2_distance", ".4f"),
    ]
    
    # Display similarity metrics
    ax3.text(0.05, y_pos, "Similarity Metrics:", fontsize=12, fontweight="bold", 
             transform=ax3.transAxes, va='top')
    y_pos -= 0.05
    
    for name, key, fmt in similarity_metrics:
        value = metrics[key]
        ax3.text(0.1, y_pos, f"{name}: {value:{fmt}}", fontsize=10, 
                 transform=ax3.transAxes, va='top')
        y_pos -= 0.05
    
    y_pos -= 0.02
    
    # Display spectral metrics
    ax3.text(0.05, y_pos, "Spectral Metrics:", fontsize=12, fontweight="bold",
             transform=ax3.transAxes, va='top')
    y_pos -= 0.05
    
    for name, key, fmt in spectral_metrics:
        value = metrics[key]
        ax3.text(0.1, y_pos, f"{name}: {value:{fmt}}", fontsize=10,
                 transform=ax3.transAxes, va='top')
        y_pos -= 0.05
    
    y_pos -= 0.02
    
    # Display distance metrics
    ax3.text(0.05, y_pos, "Distance Metrics:", fontsize=12, fontweight="bold",
             transform=ax3.transAxes, va='top')
    y_pos -= 0.05
    
    for name, key, fmt in distance_metrics:
        value = metrics[key]
        ax3.text(0.1, y_pos, f"{name}: {value:{fmt}}", fontsize=10,
                 transform=ax3.transAxes, va='top')
        y_pos -= 0.05
    
    # Add optimal lag information if meaningful
    if abs(metrics["optimal_lag"]) > 0:
        y_pos -= 0.02
        ax3.text(0.05, y_pos, f"Optimal Lag: {metrics['optimal_lag']} steps", 
                 fontsize=10, fontweight="bold", transform=ax3.transAxes, va='top')
    
    # Add overall title
    if sample_idx is not None:
        fig.suptitle(f"Attribution Map Comparison - Sample {sample_idx}", 
                     fontsize=16, fontweight="bold")
    else:
        fig.suptitle("Attribution Map Comparison", fontsize=16, fontweight="bold")
    
    plt.tight_layout()
    plt.show()


def compare_attribution_maps(original_attr: np.ndarray,
                           perturbed_attr: np.ndarray,
                           original_label: str = "Original",
                           perturbed_label: str = "Perturbed",
                           sample_idx: int = None,
                           plot: bool = True,
                           figsize: tuple = (15, 8)) -> dict:
    """
    Complete function to compute metrics and optionally plot attribution comparison.
    Uses existing functions for modular design.
    
    Parameters
    ----------
    original_attr, perturbed_attr : np.ndarray
        Attribution maps to compare.
    original_label, perturbed_label : str
        Labels for the attribution maps.
    sample_idx : int, optional
        Sample index for labeling.
    plot : bool, default=True
        Whether to create comparison plot.
    figsize : tuple
        Figure size for plot.
    
    Returns
    -------
    dict
        All computed metrics.
    """
    # Compute all metrics using the existing function
    metrics = compute_all_attribution_metrics(original_attr, perturbed_attr)
    
    # Optionally create plot
    if plot:
        plot_attribution_comparison(
            original_attr, perturbed_attr, metrics,
            original_label, perturbed_label, sample_idx, figsize
        )
    
    return metrics


def batch_attribution_comparison(idx: int, perturbations: dict, cp: int = 3) -> dict:
    """
    Compare all perturbations with original for a given sample.
    Uses existing collect_series_and_attributions function for modular design.
    
    Parameters
    ----------
    idx : int
        Sample index in x_test.
    perturbations : dict
        Perturbation parameters dictionary.
    cp : int, default=3
        Number of change points for LIME.
    
    Returns
    -------
    dict
        Metrics for each perturbation comparison.
    """
    # Use existing function to collect all data
    collected_data = collect_series_and_attributions(idx, perturbations, cp)
    original_attr = collected_data[0][2]  # Original is always first
    
    results = {}
    for i, (label, series, attr_map) in enumerate(collected_data[1:], 1):
        # Use existing comparison function without plotting
        metrics = compare_attribution_maps(
            original_attr, attr_map,
            original_label="Original",
            perturbed_label=label,
            sample_idx=idx,
            plot=False  # No plotting for batch processing
        )
        results[label] = metrics
    
    return results
```

##### ASI

```python
import numpy as np
from scipy.spatial.distance import jensenshannon
from scipy.stats import pearsonr
import pandas as pd

def calculate_asi_score(original_attr, perturbed_attr, original_ts, perturbed_ts, 
                       original_pred, perturbed_pred, weights=None):
    """Calculate ASI score for one perturbation."""
    if weights is None:
        weights = [0.25, 0.25, 0.25, 0.25]
    # Binary flip
    bf = 1.0 if np.argmax(original_pred) == np.argmax(perturbed_pred) else 0.0
    # Jensen-Shannon distance
    orig_prob = np.abs(original_pred) / np.sum(np.abs(original_pred))
    pert_prob = np.abs(perturbed_pred) / np.sum(np.abs(perturbed_pred))
    js_dist = jensenshannon(orig_prob, pert_prob)
    # Attribution correlation
    pcc_att, _ = pearsonr(original_attr.flatten(), perturbed_attr.flatten())
    pcc_att = max(-1.0, min(1.0, pcc_att)) if not np.isnan(pcc_att) else 0.0
    # Time series correlation
    pcc_ts, _ = pearsonr(original_ts.flatten(), perturbed_ts.flatten())
    pcc_ts = max(-1.0, min(1.0, pcc_ts)) if not np.isnan(pcc_ts) else 0.0
    # ASI formula (lower = better)
    asi_score = (weights[0] * (1 - bf) + weights[1] * js_dist + 
                 weights[2] * (1 - pcc_att) + weights[3] * (1 - pcc_ts))
    return {
        'asi_score': asi_score,
        'asi_binary_flip': bf,
        'asi_js_distance': js_dist,
        'asi_attribution_correlation': pcc_att,
        'asi_timeseries_correlation': pcc_ts
    }

def add_asi_to_results(full_results, sample_indices, perturbations, n_runs=10, cp=3):
    """Add ASI columns to your results DataFrame."""
    asi_results = []
    for idx in sample_indices:
        for perturbation_name, params in perturbations.items():
            for run in range(n_runs):
                # Get original data and explanation
                original_ts = x_test[idx].reshape(-1)
                original_exp = get_explanation(idx, cp=cp)
                original_attr = build_attribution_map(original_exp)
                original_pred = model.predict(original_ts[np.newaxis, :, np.newaxis])[0]
                # Apply perturbation
                perturbed_ts = apply_perturbation(idx, perturbation_name, **params)
                perturbed_exp = explain_series(perturbed_ts, cp=cp)
                perturbed_attr = build_attribution_map(perturbed_exp)
                perturbed_pred = model.predict(perturbed_ts[np.newaxis, :, np.newaxis])[0]
                # Calculate ASI
                asi_result = calculate_asi_score(
                    original_attr, perturbed_attr, original_ts, perturbed_ts,
                    original_pred, perturbed_pred
                )
                asi_results.append({
                    'sample_idx': idx,
                    'perturbation': perturbation_name,
                    'run': run,
                    **asi_result
                })
    asi_df = pd.DataFrame(asi_results)
    # Merge with your existing results
    return pd.merge(full_results, asi_df, on=['sample_idx', 'perturbation', 'run'])
```

```python
def analyze_single_perturbation(sample_idx: int, perturbation: str, params: dict, n_runs: int = 10, cp: int = 3) -> list:
    """Analyze one perturbation type on one sample with multiple runs. Returns list of result dicts for each run."""
    results = []
    # Get original explanation once
    orig_exp = get_explanation(sample_idx, cp=cp)
    orig_attr = build_attribution_map(orig_exp)
    
    for run in range(n_runs):
        # Generate perturbed series
        pert_series = apply_perturbation(sample_idx, perturbation, **params)
        
        # Compute perturbed explanation
        pert_exp = explain_series(pert_series, cp=cp)
        pert_attr = build_attribution_map(pert_exp)
        
        # Compute metrics
        metrics = compute_all_attribution_metrics(orig_attr, pert_attr)
        
        results.append({
            "sample_idx": sample_idx,
            "perturbation": perturbation,
            "run": run,
            **metrics
        })
    
    return results

def analyze_sample(sample_idx: int, perturbations: dict, n_runs: int = 10, cp: int = 3) -> pd.DataFrame:
    """ Analyze all perturbations on one sample. Returns DataFrame with results for all perturbations and runs."""
    all_results = []
    for p_name, params in perturbations.items():
        p_results = analyze_single_perturbation(
            sample_idx, p_name, params, n_runs, cp
        )
        all_results.extend(p_results)
    
    return pd.DataFrame(all_results)

def run_full_analysis(sample_indices: list, perturbations: dict, n_runs: int = 10, cp: int = 3) -> pd.DataFrame:
    """Run full analysis pipeline for multiple samples. Returns combined DataFrame of all results."""
    all_dfs = []
    for idx in tqdm(sample_indices, desc="Processing samples"):
        sample_df = analyze_sample(idx, perturbations, n_runs, cp)
        all_dfs.append(sample_df)
    
    return pd.concat(all_dfs, ignore_index=True)


def plot_summary_statistics(summary_stats):
    # Plot settings
    fig, axs = plt.subplots(3, 2, figsize=(16, 12))
    axs = axs.flatten()
    metrics = summary_stats.columns.levels[0]

    for i, metric in enumerate(metrics):
        ax = axs[i]
        df_metric = summary_stats[metric]
        bars = ax.bar(
            summary_stats.index,
            df_metric["mean"],
            yerr=df_metric["std"],
            capsize=5
        )
        ax.set_title(f"{metric.replace('_', ' ').title()} (mean ± std)")
        ax.set_ylabel("Score")
        ax.set_xticks(range(len(summary_stats.index)))
        ax.set_xticklabels(summary_stats.index, rotation=45)

    fig.suptitle("Summary Statistics for Different Perturbation Methods", fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
