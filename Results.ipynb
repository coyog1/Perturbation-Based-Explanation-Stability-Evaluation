```python
def flatten_cols(df):
    new = []
    for col in df.columns:
        m = re.match(r"\('(.+)',\s*'(.+)'\)", col)
        if m:
            new.append(f"{m.group(1)}_{m.group(2)}")
        else:
            new.append(col)
    df.columns = new

flatten_cols(summary_stats_lstm_forda)
flatten_cols(summary_stats_lstm_fordb)
```

### Consistency Heatmaps

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Assemble your model summary-stat DataFrames
model_dfs = {
    'CNN–FordA': summary_stats_cnn_forda,
    'CNN–FordB': summary_stats_cnn_fordb,
    'LSTM–FordA': summary_stats_lstm_forda,
    'LSTM–FordB': summary_stats_lstm_fordb
}

# 2. Define perturbations (rows) and metrics (columns)
perturbations = ['amplitude_scale', 'contiguous_mask', 'frequency_mask', 'phase_shift', 'value_mask']
metrics = [
    'circular_cross_correlation',
    'cosine_similarity',
    'combined_spectral_score',
    'fourier_distance',
    'earth_mover_distance'
]

# 3. Classify metrics by direction
benefit_metrics = {'circular_cross_correlation', 'cosine_similarity', 'combined_spectral_score'}

# 4. Create a 2×2 grid of heatmaps
fig, axes = plt.subplots(2, 2, figsize=(12, 10), dpi=100)
axes = axes.flatten()

for ax, (model_name, df) in zip(axes, model_dfs.items()):
    # Build a DataFrame of reliability scores
    reliab = pd.DataFrame(index=perturbations, columns=metrics, dtype=float)
    for metric in metrics:
        mu    = df[f'{metric}_mean']
        sigma = df[f'{metric}_std']
        cv    = (sigma.abs() / mu.abs()).clip(upper=1)
        if metric in benefit_metrics:
            score = mu * (1 - cv)
        else:
            score = (1 / (1 + mu)) * (1 - cv)
        reliab[metric] = score.values
    
    # Plot heatmap
    sns.heatmap(
        reliab, annot=True, fmt='.2f', cmap='viridis',
        cbar_kws={'label': 'Reliability'}, ax=ax
    )
    ax.set_title(f'{model_name}', fontsize=14)
    ax.set_xlabel('')
    ax.set_ylabel('Perturbation Method')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_metric_heatmaps(summary_stats: pd.DataFrame, model_name: str):
    """
    Given a DataFrame `summary_stats` with columns:
      - '{metric}_mean'
      - '{metric}_std'
    for each metric in ['circular_cross_correlation', 'cosine_similarity',
                       'combined_spectral_score', 'fourier_distance', 'earth_mover_distance']
    and an index of perturbations ['amplitude_scale', 'contiguous_mask', 
    'frequency_mask', 'phase_shift', 'value_mask'],
    this function will generate three sequential heatmaps:
      1. Consistency (1 - CV)
      2. Sensitivity (normalized Δμ)
      3. Interpretability (Consistency × Sensitivity)
    """
    perturbations = [
        'amplitude_scale',
        'contiguous_mask',
        'frequency_mask',
        'phase_shift',
        'value_mask'
    ]
    metrics = [
        'circular_cross_correlation',
        'cosine_similarity',
        'combined_spectral_score',
        'fourier_distance',
        'earth_mover_distance'
    ]
    benefit_metrics = {
        'circular_cross_correlation',
        'cosine_similarity',
        'combined_spectral_score'
    }

    df = summary_stats.copy()

    # 1) Consistency
    consistency = pd.DataFrame(index=perturbations, columns=metrics, dtype=float)
    for m in metrics:
        mu = df[f'{m}_mean']
        sigma = df[f'{m}_std']
        cv = (sigma.abs() / mu.abs()).clip(upper=1)
        consistency[m] = (1 - cv).values

    plt.figure(figsize=(7, 6), dpi=100)
    sns.heatmap(
        consistency,
        annot=True,
        fmt='.2f',
        cmap='RdYlGn',
        cbar_kws={'label': 'Consistency (1 − CV)'}
    )
    plt.xticks(
        ticks=[i + 0.5 for i in range(len(metrics))],
        labels=['CCC', 'Cosine', 'Spectral', 'Fourier', 'EMD'],
        rotation=0,
        ha='center'
    )
    plt.yticks(
        ticks=[i + 0.5 for i in range(len(perturbations))],
        labels=['Amp\nScale', 'Contiguous', 'Frequency', 'Phase\nShift', 'Value\nMask'],
        rotation=0,
        ha='right'
    )
    plt.title(f'{model_name}: Consistency', fontsize=14, fontweight='bold')
    plt.xlabel('Metric')
    plt.ylabel('Perturbation')
    plt.tight_layout()
    plt.show()

    # 2) Sensitivity
    sensitivity = pd.DataFrame(index=perturbations, columns=metrics, dtype=float)
    for m in metrics:
        mu = df[f'{m}_mean'].values
        if m in benefit_metrics:
            raw = mu.max() - mu
        else:
            raw = mu - mu.min()
        if raw.max() > 0:
            norm = raw / raw.max()
        else:
            norm = raw * 0
        sensitivity[m] = norm

    plt.figure(figsize=(7, 6), dpi=100)
    sns.heatmap(
        sensitivity,
        annot=True,
        fmt='.2f',
        cmap='RdYlGn',
        cbar_kws={'label': 'Sensitivity'}
    )
    plt.xticks(
        ticks=[i + 0.5 for i in range(len(metrics))],
        labels=['CCC', 'Cosine', 'Spectral', 'Fourier', 'EMD'],
        rotation=0,
        ha='center'
    )
    plt.yticks(
        ticks=[i + 0.5 for i in range(len(perturbations))],
        labels=['Amp\nScale', 'Contiguous', 'Frequency', 'Phase\nShift', 'Value\nMask'],
        rotation=0,
        ha='right'
    )
    plt.title(f'{model_name}: Sensitivity', fontsize=14, fontweight='bold')
    plt.xlabel('Metric')
    plt.ylabel('Perturbation')
    plt.tight_layout()
    plt.show()

    # 3) Interpretability
    interpretability = consistency * sensitivity
    
    plt.figure(figsize=(7, 6), dpi=100)
    sns.heatmap(
        interpretability,
        annot=True,
        fmt='.2f',
        cmap='RdYlGn',  # red-to-green colormap
        cbar_kws={'label': 'Interpretability (Consistency × Sensitivity)'}
    )
    plt.xticks(
        ticks=[i + 0.5 for i in range(len(metrics))],
        labels=['CCC', 'Cosine', 'Spectral', 'Fourier', 'EMD'],
        rotation=0,
        ha='center'
    )
    plt.yticks(
        ticks=[i + 0.5 for i in range(len(perturbations))],
        labels=['Amp\nScale', 'Contiguous', 'Frequency', 'Phase\nShift', 'Value\nMask'],
        rotation=0,
        ha='right'
    )
    plt.title(f'{model_name}: Interpretability Scores', fontsize=14, fontweight='bold')
    plt.xlabel('Metric')
    plt.ylabel('Perturbation')
    plt.tight_layout()
    plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define models and their data
model_dfs = {
    'CNN–FordA': summary_stats_cnn_forda,
    'CNN–FordB': summary_stats_cnn_fordb,
    'LSTM–FordA': summary_stats_lstm_forda,
    'LSTM–FordB': summary_stats_lstm_fordb
}

# Define full perturbation names (must match order in summary_stats)
perturbations = [
    'amplitude_scale',
    'contiguous_mask',
    'frequency_mask',
    'phase_shift',
    'value_mask'
]

# Define masking perturbations
masking_perturbations = {'value_mask', 'contiguous_mask', 'frequency_mask'}

# Define metrics
metrics = [
    'circular_cross_correlation',
    'cosine_similarity',
    'combined_spectral_score',
    'fourier_distance',
    'earth_mover_distance'
]
benefit_metrics = {'circular_cross_correlation', 'cosine_similarity', 'combined_spectral_score'}

# Replace index in each model DataFrame
for df in model_dfs.values():
    df.index = perturbations

# Compute average detection score for masking perturbations
masking_scores = {}
for model_name, df in model_dfs.items():
    detection_vals = []
    for metric in metrics:
        mu    = df.get(f'{metric}_mean')
        sigma = df.get(f'{metric}_std')
        if mu is None or sigma is None:
            continue
        cv = (sigma.abs() / mu.abs()).clip(upper=1)
        raw_score = mu * (1 - cv) if metric in benefit_metrics else (1 / (1 + mu)) * (1 - cv)
        detection_score = 1 - raw_score
        for pert in masking_perturbations:
            if pert in detection_score.index:
                detection_vals.append(detection_score.loc[pert])
    if detection_vals:
        masking_scores[model_name] = np.mean(detection_vals)
    else:
        print(f"⚠️  No valid masking perturbations found for model: {model_name}")
        masking_scores[model_name] = np.nan

# Plot
import matplotlib.pyplot as plt

# Custom colors
colors = ['#1f4e79', '#6baed6', '#2ca02c', '#98df8a']  # Dark blue, light blue, dark green, light green

plt.figure(figsize=(8, 4.5))

bars = plt.bar(
    masking_scores.keys(),
    masking_scores.values(),
    color=colors,
    edgecolor='black',
    linewidth=1.0
)

# Grid and aesthetics
plt.grid(axis='y', linestyle='--', linewidth=0.4, color='gray', alpha=0.6)
plt.ylim(0, 1)
plt.ylabel('', fontsize=11)
plt.title('Average Detection Score (Masking Perturbations)', fontsize=13, fontweight='bold')
plt.xticks(rotation=0, ha='center', fontsize=11)

# Add spacing and tighten layout
plt.subplots_adjust(left=0.12, right=0.95, top=0.88, bottom=0.12)
plt.tight_layout()
plt.show()

```
